# Model Inference Output

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import torch
import torch.nn as nn
from unet_vgg4_cc import UnetVggCC

def heatmap2d(arr: np.ndarray):
    plt.imshow(arr, cmap='viridis')
    plt.colorbar()
    plt.show()
```

```python
root = "./datasets/agriculture/train_data/images/img"
models_path = './checkpoints/agriculture_cpt/epoch_99.pth'

gpu_or_cpu='cuda'
dropout_keep_prob = 1.0
initial_pad = 126
interpolate = 'False'
conv_init = 'he'
n_classes = 1
n_channels = 1

device = torch.device(gpu_or_cpu)
model = UnetVggCC(kwargs={'dropout_keep_prob':dropout_keep_prob, 'initial_pad':initial_pad, 'interpolate':interpolate, 'conv_init':conv_init, 'n_classes':n_classes, 'n_channels':n_channels})
# model.cuda(0)
criterion_sig = nn.Sigmoid()
model.load_state_dict(torch.load(models_path, map_location=device), strict=True)
# model.cuda(0)
```

```python
img = plt.imread(os.path.join(root, "0174_4.jpg")) / 255
img_gpu = img.transpose((2,0,1))
img_gpu = torch.from_numpy(img_gpu)
img_gpu = img_gpu.cuda(0)
img.shape
```
